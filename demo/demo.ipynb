```python
if data['dominant_eye'] ==0:
    fix = data['left_eye']
else:
    fix = data['right_eye']
```

```python
s = im.shape
N1 = len(fix)
N1 = int(N1/4)
fix_map = np.zeros((s[0],s[1]))
for j in range(N1-3):
    temp = fix[j*4+10]
    if temp and temp[0]>0 and temp[0]<1 and temp[1]>0 and temp[1]<1:

        x = int(temp[0]*s[1])
        y = int(temp[1]*s[0])
        fix_map[y,x] = 1
```

```python
##show the smoothed fixation map (saliency map)
plt.imshow(ndimage.filters.gaussian_filter(fix_map, 19))
```
